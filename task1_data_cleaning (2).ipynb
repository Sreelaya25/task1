```python
import pandas as pd

# Load CSV with tab separator
df = pd.read_csv("/content/marketing_campaign.csv", sep='\t')

# Clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Preview
df.head()
```

```python
# Check missing values
print(df.isnull().sum())

# Drop rows with missing values (or choose to fillna if needed)
df = df.dropna()
```

```python
df = df.drop_duplicates()
```

```python
# Clean string columns
df['education'] = df['education'].str.lower().str.strip()
df['marital_status'] = df['marital_status'].str.lower().str.strip()
```

```python
df['dt_customer'] = pd.to_datetime(df['dt_customer'], dayfirst=True)
```

```python
# Convert income to integer (if float)
df['income'] = df['income'].astype(int)
```

```python
df.to_csv('/content/marketing_campaign_cleaned.csv', index=False)
```
